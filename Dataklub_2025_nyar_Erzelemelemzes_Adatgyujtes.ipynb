```python
import requests
import pandas as pd

# web scraping
from bs4 import BeautifulSoup
import re

#AI
import json
#import openai
from io import StringIO
```

```python
nagy_kat = { "poz" :['Szeretet','Boldogsag' ,'Bizalom', 'Mosoly'] , "neg" :['Félelem','Siras' ,'Fajdalom', 'Szenvedes']}
```

```python
def kigyujt_versek(soup, kulcs):
    adatok =[]
    for div in soup.find_all("div", id=re.compile(r'^\d+$')):
        verscim = div.find("p", class_="verscim")
        szerzo = div.find("p", class_="auth")
        more_link = div.find("p", class_="more")
        a_tag = more_link.find("a") if more_link else None

        if verscim and szerzo and a_tag:
            adat = {
                "id": div["id"],
                "szerzo": szerzo.get_text(strip=True),
                "cim": verscim.get_text(strip=True),
                "link": a_tag.get("href"),
                "tipus": kulcs
            }
            adatok.append(adat)

    return adatok
```

```python

#Tanító és Teszt adatok (tanító és teszt versek) legyűjtése
kat_oldalak = []
for kulcs, lista in nagy_kat.items():
  for  item  in lista:
    for i in range(1, 3):
        url = "https://poet.hu/kategoria/"+item+"/"+str(i)
        #print(url)
        headers = {'User-Agent': 'Mozilla/5.0'}
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'lxml')
        kat_oldalak.append(kigyujt_versek(soup, kulcs))
```

```python
# vers saját oldalán a vers szövege kiszed
def vers_kiszed(soup):
    vers_elem = soup.find(class_="vers")
    if vers_elem:
      #vers_szoveg = vers_elem.get_text(strip=True, separator="\n")
      vers_szoveg = vers_elem.get_text(strip=True, separator=" ")
      # sortörések eltávolítása, ha maradt
      vers_szoveg = " ".join(vers_szoveg.split())
      #print("A vers szövege:\n")
      #print(vers_szoveg)
      return vers_szoveg
```

```python
vers_info = []
for  oldal in kat_oldalak:
  for  elem in oldal:
   print(elem )
   if isinstance(elem, dict) and 'link' in elem :
    url2 = "https://poet.hu/"+elem['link']
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url2, headers=headers)
    soup = BeautifulSoup(response.text, 'lxml')
    vers_info.append({'tipus': elem['tipus'] ,'szerzo': elem['szerzo'] , 'cim': elem['cim'], 'vers':vers_kiszed(soup), 'link': 'https://poet.hu' + elem['link'] })
```

```python
df = pd.DataFrame(vers_info)
print(df)
```
